```python
crash_df=sns.load_dataset('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/car_crashes.csv')
```

```python
car_df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/car_crashes.csv')
```

```python
car_df.head(5)
```

```python
#creating not distribution plot for not_distracted
#bins are like bucket which distributed data.

sns.distplot(car_df['not_distracted'],kde=False,bins=25)
```

```python
sns.distplot(car_df['not_distracted'])
```

```python
#creating regression plot
sns.jointplot(x='speeding',y='alcohol',data=car_df,kind='reg')
```

```python

#creating kde plot

sns.jointplot(x='speeding',y='alcohol',data=car_df,kind='kde')
```

```python
sns.jointplot(x='speeding',y='alcohol',data=car_df,kind='hex')
```

```python
#creating kde plot for alcohol usage

sns.kdeplot(car_df['alcohol'])
```

```python
#pairplot to create all kind of combination graphs

sns.pairplot(car_df)
```
