```python
#importing required library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing the data
cs_df=pd.read_csv('https://raw.githubusercontent.com/derekbanas/pandas-tutorial/master/ComputerSales.csv')
```

```python
#getting top five rows
cs_df.head(5)
```

```python
#distribution plot in seaborn
sns.distplot(cs_df['Age'],kde=False,bins=25)
```

```python
car_df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/car_crashes.csv')
```

```python
car_df.head(5)
```

```python
#creating not distribution plot for not_distracted
#bins are like bucket which distributed data.

sns.distplot(car_df['not_distracted'],kde=False,bins=25)
```

```python
sns.distplot(car_df['not_distracted'])
```

```python
#creating regression plot
sns.jointplot(x='speeding',y='alcohol',data=car_df,kind='reg')
```

```python

#creating kde plot

sns.jointplot(x='speeding',y='alcohol',data=car_df,kind='kde')
```

```python
sns.jointplot(x='speeding',y='alcohol',data=car_df,kind='hex')
```

```python
#creating kde plot for alcohol usage

sns.kdeplot(car_df['alcohol'])
```

```python
#pairplot to create all kind of combination graphs

sns.pairplot(car_df)
```

```python
tip_df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```python
tip_df.head(5)
```

```python
# following is categorical data

sns.pairplot(tip_df,hue='sex',palette='Blues')
```

```python
# rugplot creating graph from same column of data 
# which are more common are with dark blue color and which are with light blue are less common ins the 
#column
sns.rugplot(tip_df['tip'])
```

```python
sns.set_style('darkgrid')

sns.jointplot(x='speeding',y='alcohol',data=car_df,kind='reg')
```

```python
sns.set_style('ticks')
plt.figure(figsize=(8,4))
sns.set_context('paper',font_scale=1.4)


sns.jointplot(x='speeding',y='alcohol',data=car_df,kind='reg')
```

```python
#categorical data with seaborn
sns.barplot(x='sex',y='total_bill',data=tip_df)
```

```python
#categorical data of mean value with seaborn
sns.barplot(x='sex',y='total_bill',data=tip_df,estimator=np.mean)
```

```python
#categorical data with seaborn
sns.barplot(x='sex',y='total_bill',data=tip_df,estimator=np.median)
```

```python
#categorical data with seaborn
sns.barplot(x='sex',y='total_bill',data=tip_df,estimator=np.std)
```

```python
#categorical data with seaborn
sns.barplot(x='sex',y='total_bill',data=tip_df,estimator=np.var)
```

```python
#categorical data with seaborn
sns.barplot(x='sex',y='total_bill',data=tip_df,estimator=np.cov)
```

```python
#counting the sex with countplot
sns.countplot(x='sex',data=tip_df)
```

```python
#getting categorical data on multiple day

sns.boxplot(x='day',y='total_bill',data=tip_df,hue='sex')
plt.legend(loc=0)
```

```python
#creating violinplot
sns.violinplot(x='day',y='total_bill',data=tip_df,hue='sex')
```

```python
#creating stripplot
sns.stripplot(x='day',y='total_bill',data=tip_df)
```

```python
#creating stripplot
sns.stripplot(x='day',y='total_bill',data=tip_df,hue='sex',dodge='True')
```

```python
sns.violinplot(x='day',y='total_bill',data=tip_df,color='white')
```
